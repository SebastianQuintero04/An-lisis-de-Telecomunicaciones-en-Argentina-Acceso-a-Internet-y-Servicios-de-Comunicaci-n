## Gráfico 1: Histograma de 'Accesos por cada 100 hogares'

El primer gráfico es un histograma que muestra la distribución de los accesos a internet por cada 100 hogares. Este gráfico nos ayudará a visualizar cómo se distribuyen los accesos a internet en la población y si existen patrones o valores atípicos.

# Cálculo y visualización de KPIs para accidentes de tránsito en CABA

 ## Introducción
En este estudio, exploraremos y calcularemos varios KPIs fundamentales para evaluar la infraestructura y accesibilidad de Internet en diferentes provincias de Argentina. Utilizaremos datos detallados de múltiples datasets para analizar tres KPIs específicos:

KPI de Mejora Necesaria en Acceso a Internet:
Este indicador nos permitirá entender cuánto necesita mejorar cada provincia para alcanzar un objetivo de tasa de cambio del 2% en el acceso a Internet. Calcularemos este KPI trimestralmente utilizando datos del dataset 1, comparando la tasa de cambio actual con el objetivo establecido.

KPI de Distribución de Tecnologías de Acceso por Provincia:
En este análisis, evaluaremos la disponibilidad y penetración de la fibra óptica en diferentes localidades utilizando datos combinados de los datasets 1. Este KPI nos ayudará a medir la extensión de la cobertura y la intensidad de uso de este servicio crucial de comunicaciones.

KPI de Proporción de Población con Acceso a Fibra Óptica por Provincia>
Evaluaremos la disponibilidad y penetración de la fibra óptica en diferentes localidades utilizando datos combinados de el data set 2. Este KPI nos ayudará a medir la extensión de la cobertura y la intensidad de uso de este servicio crucial de comunicaciones.

Estos KPIs nos permitirán realizar un análisis exhaustivo de la infraestructura de telecomunicaciones en Argentina, destacando áreas de mejora y áreas donde las metas ya han sido superadas. Además, propondremos y mediremos otros indicadores relevantes para enriquecer nuestro entendimiento de este crucial aspecto de la conectividad en el país.

 ## KPI 1 de Mejora Necesaria en Acceso a Internet:


```python
plt.style.use('seaborn-v0_8-dark-palette')
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
# KPI 1: Mejora Necesaria en Acceso a Internet
def calcular_kpi_mejora(df):
    df['Tasa_Cambio_Anual'] = df.groupby('Provincia')['Accesos por cada 100 hogares'].pct_change(4)
    objetivo = 0.02
    df['Mejora_Necesaria'] = objetivo - df['Tasa_Cambio_Anual']
    return df.groupby('Provincia')['Mejora_Necesaria'].mean()

kpi_mejora = calcular_kpi_mejora(df_internet)

# Visualización KPI 1
plt.figure(figsize=(14, 8))
kpi_mejora.sort_values().plot(kind='bar')
plt.title('KPI 1: Mejora Necesaria en Acceso a Internet por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Mejora Necesaria (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```


## KPI 2: Distribución de Tecnologías de Acceso por Provincia


```python
def calcular_kpi_distribucion_tecnologias(df_internet):
    # Seleccionar las columnas relevantes para calcular la proporción de cada tecnología
    tecnologias_cols = ['Porcentaje_ADSL', 'Porcentaje_Cablemodem', 'Porcentaje_Fibra óptica', 
                        'Porcentaje_Wireless', 'Porcentaje_Otros']
    
    # Calcular la proporción de cada tecnología por provincia
    proporciones_tecnologias = df_internet.groupby('Provincia')[tecnologias_cols].mean()
    
    # Calcular el KPI de distribución de tecnologías como la desviación estándar de las proporciones
    kpi_distribucion_tecnologias = proporciones_tecnologias.std(axis=1)
    
    return kpi_distribucion_tecnologias

# Calcular el KPI de distribución de tecnologías
kpi_distribucion_tecnologias = calcular_kpi_distribucion_tecnologias(df_internet)

# Mostrar resultados
print("KPI de Distribución de Tecnologías de Acceso por Provincia:")
print(kpi_distribucion_tecnologias)
```

```python
# Visualización del KPI de Distribución de Tecnologías de Acceso por Provincia
plt.figure(figsize=(12, 8))
kpi_distribucion_tecnologias.sort_values().plot(kind='barh', color='skyblue')
plt.title('KPI de Distribución de Tecnologías de Acceso por Provincia')
plt.xlabel('Desviación Estándar de Proporciones')
plt.ylabel('Provincia')
plt.grid(axis='x', linestyle='--', alpha=0.8)
plt.tight_layout()
plt.show()
```

## KPI 3: Velocidad Media de Bajada

```python
# Función para calcular el KPI de proporción de población con fibra óptica
def calcular_kpi_fibra_optica(df_mapa):
    df_mapa['Proporcion_Fibra_Optica'] = df_mapa['Fibra óptica'] / df_mapa['Población']
    kpi_fibra_optica = df_mapa.groupby('Provincia')['Proporcion_Fibra_Optica'].mean()
    return kpi_fibra_optica

# Calcular el KPI de fibra óptica usando solo el dataset 2 (df_mapa)
kpi_fibra_optica = calcular_kpi_fibra_optica(df_mapa)

# Visualización del KPI de proporción de población con acceso a fibra óptica
plt.figure(figsize=(12, 6))
kpi_fibra_optica.sort_values(ascending=False).plot(kind='bar', color='skyblue')
plt.title('Proporción de Población con Acceso a Fibra Óptica por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Proporción de Población con Fibra Óptica')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
 #Mapa de calor para comparar los tres KPIs
kpis_combinados = pd.DataFrame({
    'Mejora Necesaria': kpi_mejora,
    'Cobertura Fibra Óptica': kpi_fibra_optica,
    'Distribucion Tecnologias': kpi_distribucion_tecnologias
    })
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(kpis_combinados, annot=True, cmap='YlOrRd', fmt='.2f')
plt.title('Comparación de KPIs por Provincia')
plt.tight_layout()
plt.show()
```
